```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()

# # create a very large dataframe with random values

df['a'] = np.random.randint(0, 100, 50000000)
df['b'] = np.random.randint(0, 1000, 50000000)
```

```python
df.to_csv('a.csv', index=False)
```

```python
# size fo the dataframe in mb
df.memory_usage(index=False).sum() / 1024 / 1024
```
