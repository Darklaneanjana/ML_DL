```python
import spacy
#English core package
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("Dr. sean some paragraph about the customer complaint. I ate an apple")
```

```python
for i in doc.sents:
    print(i)
for i in doc.sents:
    for j in i:
        print(j)
```
